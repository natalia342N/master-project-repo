```python
import subprocess # using copilot for plotting
import numpy as np
import matplotlib.pyplot as plt


executable = "./csmca.py"  
cuda_files = {
    "Pipelined CG": "pip-cg3.cu",
    "Classical CG": "classical-cg.cu"
}

results = {label: {"unknowns": [], "time_per_iteration": []} for label in cuda_files.keys()}

# Run 
for label, cuda_file in cuda_files.items():
    print(f"Running the C++ program for {label} ({cuda_file})...")
    try:
        result = subprocess.run(
            [executable, cuda_file],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )

        output = result.stdout
        lines = output.split("\n")
        current_unknowns = None

        for line in lines:
            if "Solving Ax=b with" in line:
                current_unknowns = int(line.split("with")[1].split("unknowns")[0].strip())

            elif "Time per iteration" in line and current_unknowns is not None:
                iteration_time = float(line.split(":")[1].strip().split()[0])  
                results[label]["unknowns"].append(current_unknowns)
                results[label]["time_per_iteration"].append(iteration_time)
                current_unknowns = None 

        print(f"Parsed Data for {label}:")
        print("Unknowns:", results[label]["unknowns"])
        print("Time per Iteration:", results[label]["time_per_iteration"])

    except Exception as e:
        print(f"Error running the C++ program for {label}: {e}")


for label, data in results.items():
    unknowns = np.array(data["unknowns"])
    time_per_iteration = np.array(data["time_per_iteration"])
    if len(unknowns) == len(time_per_iteration):  
        plt.loglog(unknowns, time_per_iteration, marker='o', label=f'{label} Time per Iteration')


plt.xlabel('Number of Unknowns N')
plt.ylabel('Time per Iteration seconds')
plt.title('Time per Iteration vs Number of Unknowns ')
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
